## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
_______________________________________________________________________________________________________________________________
A Primary Key is a unique identifier for each record in a table, which ensures that there are no two rows that have the same value in that particular column.

A Foreign Key is a column in one table that links to the Primary Key in another table,which establishes a relationship between two tables.




## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
One-to-One relationship shows that each record in one table is related to a single record in another table.

One-to-Many relationship shows a single record in one table can be related to multiple records in another table. 

Many-to-One relationship entails multiple records in one table that can relate to a single record in another. This is essentially the reverse of a one-to-many relationship.

Many-to-Many relationship shows records in one table can relate to multiple records in another table and vice versa. 

## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
Inner join returns records with matching values in both tables.
Left join returns all records from the left table and matched records from the right; if no match, NULLs are returned.
Right join returns all records from the right table and matched records from the left; if no match, NULLs are returned.
Full join combines results of both left and right joins, returning all records from both tables with NULLs where there are no matches.

Aggregate functions perform calculations on a set of values to return a single summary value (e.g., SUM, AVG, COUNT), while the GROUP BY statement is used to arrange identical data into groups. However, when used together, GROUP BY organizes the result set into groups based on one or more columns, and aggregate functions calculate values for each group. 

## Question 6

What role do Associative Entities play (JOIN Tables) in many-to-many JOINs?


Type your answer below this line:
_______________________________________________________________________________________________________________________________
Associative entities, or join tables, play a crucial role in many-to-many relationships in SQL databases. They serve as a bridge between two entities by holding foreign keys that reference the primary keys of both entities involved. This structure allows for the creation of associations between multiple records in each table.
